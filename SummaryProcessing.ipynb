```python
#Process for Boxplots
import pandas as pd
import random
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt 
import math
```

#### 1 Day Run Values

```python

#1 day interval
numVehicles=22

dfSummaryValues = pd.read_csv("Output/MonthRun1dayInterval2/SummaryValues.csv") #income for each  sim run
dfDegradation = pd.read_csv("Output/MonthRun1dayInterval2/Degradation.csv") #degradation cost, 31 days x 22 vehicles
dfCycles = pd.read_csv("Output/MonthRun1dayInterval2/Cycles.csv") #representative solar day, 15 min interval
dfCharging=pd.read_csv("Output/MonthRun1dayInterval2/Charging.csv")


arrSOCs=np.genfromtxt("Output/MonthRun1dayInterval2/arrSOCs.csv",delimiter=",")
arrSOCs=np.reshape(arrSOCs[22:,:],(22,2880))
arrRegDown=np.genfromtxt("Output/MonthRun1dayInterval2/arrRegDown.csv",delimiter=",")
arrRegDown=np.reshape(arrRegDown[22:,:],(22,2880))
arrRegDownDispatch=np.genfromtxt("Output/MonthRun1dayInterval2/arrRegDownDispatch.csv",delimiter=",")
arrRegDownDispatch=np.reshape(arrRegDownDispatch[22:,:],(22,2880))
arrRegUp=np.genfromtxt("Output/MonthRun1dayInterval2/arrRegUp.csv",delimiter=",")
arrRegUp=np.reshape(arrRegUp[22:,:],(22,2880))
arrRegUpDispatch=np.genfromtxt("Output/MonthRun1dayInterval2/arrRegUpDispatch.csv",delimiter=",")
arrRegUpDispatch=np.reshape(arrRegUpDispatch[22:,:],(22,2880))
    
fig, ax = plt.subplots(figsize=(11,7))
for ind in range(1,np.shape(arrRegDownDispatch)[1]):

    if ind == 1: 
        
        ax.fill_between(np.arange(96*30),np.zeros(96*30),np.array(arrRegDownDispatch[ind,:]).flatten())
    
    else: 
            
        lb = np.sum(arrRegDownDispatch[0:ind,:],axis=0)
        ub = np.sum(arrRegDownDispatch[0:ind+1,:],axis=0)
       
        ax.fill_between(np.arange(96*30),np.array(lb).flatten(),np.array(ub).flatten())


plt.ylabel('kW')
plt.xlabel("Hour")
plt.title("Regulation Down Dispatch")
plt.savefig("Output/MonthRun1dayInterval2/regDownDispatch.png")
plt.close()
    
fig, ax = plt.subplots(figsize=(11,7))
for ind in range(1,np.shape(arrRegUpDispatch)[1]):

    if ind == 1: 
        
        ax.fill_between(np.arange(96*30),np.zeros(96*30),np.array(arrRegUpDispatch[ind,:]).flatten())
    
    else: 
            
        lb = np.sum(arrRegUpDispatch[0:ind,:],axis=0)
        ub = np.sum(arrRegUpDispatch[0:ind+1,:],axis=0)
       
        ax.fill_between(np.arange(96*30),np.array(lb).flatten(),np.array(ub).flatten())


plt.ylabel('kW')
plt.xlabel("Hour")
plt.title("Regulation Up Dispatch")
plt.savefig("Output/MonthRun1dayInterval2/regUpDispatch.png")
plt.close()
```
